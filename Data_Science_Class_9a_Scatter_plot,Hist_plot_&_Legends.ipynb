# Scatter Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'o',color='red')
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'-o',color='red')
plt.show()
```

```python
x= np.random.randn(100)
y= np.random.randn(100)
colors= np.random.randn(100)
sizes= 1000 * np.random.randn(100)

plt.scatter(x,y,c = colors ,s = sizes,alpha=0.5)
plt.colorbar()
plt.show()

# randn mtlb normal distribution
```

```python
data = x
plt.hist(data)
plt.show()
```

```python
plt.hist(data ,bins=30,alpha=0.5,color='green',histtype='stepfilled')
plt.show()

# bins mtlb kitna spread kia hua graph ko..jitne bins hogay utne spread krke values dikhayga
# histtype mtlb type of histogram 
```

```python
x1=np.random.normal(0,0.8,1000)
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(3,2,1000)

dict_extra = dict(histtype='stepfilled',alpha=0.3,bins=30)

plt.hist(x1, **dict_extra)
plt.hist(x2, **dict_extra)
plt.hist(x3, **dict_extra)

plt.show()
```

```python
x= np.linspace(0,10,1000)
plt.plot(x,np.sin(x), '-b' , label='Sine graph')
plt.plot(x,np.cos(x), '--r' , label='Cosine graph')

plt.legend(loc='lower left',ncol=2)
# plt.legend(loc='best')
plt.show()
```

```python
x,y = np.random.multivariate_normal([0,0],[[1,1],[1,2]],10000).T
plt.hist2d(x,y,bins=20,cmap='Reds')
plt.show()
```
